### iqr로 이상치 제거

```python
d_q3 = df['trip_distance'].quantile(0.75)
d_q1 = df['trip_distance'].quantile(0.25)

d_iqr = d_q3 - d_q1

t_q3 = df['trip_time_in_secs'].quantile(0.75)
t_q1 = df['trip_time_in_secs'].quantile(0.25)

t_iqr = t_q3 - t_q1
```

```python
print(d_q3, d_q1, d_iqr)
print(t_q3, t_q1, t_iqr)
```

```python
import numpy as np

def defect_outliers(df, features):
    outlier_indices = []
    for col in features:
        q1 = np.percentile(df[col], 25)
        q3 = np.percentile(df[col], 75)
        
        iqr = q3 - q1
        outlier_step = 1.5 * iqr
        
        outlier_list_col = df[(df[col] < q1 - outlier_step) | (df[col] > q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = set(outlier_indices)
    outlier_indices = list(outlier_indices)
    return outlier_indices
```

### trip_distance, trip_time_in_secs, fare_amount 제거 => new_df

```python
outliers_to_drop = defect_outliers(df, ['trip_distance', 'trip_time_in_secs', 'fare_amount'])
outliers_to_drop.sort()
# outliers_to_drop
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
```

```python
minmax_scaler = MinMaxScaler()
x_train = minmax_scaler.fit_transform(train_dataset)
x_train
```

### Sequential

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation = 'relu', input_shape=[x_train.shape[1]]), 
        layers.Dense(64, activation = 'relu'),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss = 'mse',
                 optimizer = optimizer,
                 metrics = ['mae', 'mse', 'accuracy'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

### 모델 훈련

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
```

```python
EPOCHS = 30
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(
  x_train, y_train,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot(), early_stop])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
hist['rmse'] = np.sqrt(hist['mse'])
hist.tail()
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(3,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
#   plt.ylim([0,0.4])
  plt.legend()

  plt.subplot(3,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
#   plt.ylim([0,0.3])
  plt.legend()
  plt.show()
    
  plt.subplot(3,1,3)
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.plot(hist['epoch'], hist['loss'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_loss'],
           label = 'Val Error')
#   plt.ylim([0,0.3])
  plt.legend()
  plt.show()

plot_history(history)
```

```python
loss, mae, mse, accuracy = model.evaluate(x_test, y_test, verbose=2)

print("테스트 세트의 평균 절대 오차: {:5.2f} MPG".format(mae))
```
