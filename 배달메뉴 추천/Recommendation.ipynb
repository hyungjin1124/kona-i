```python
import re
```

```python
def add_gu(addr):
    p = re.compile(r"(인천광역시 |인천 )(([ㄱ-ㅣ가-힣]+\d*동[^구])|([ㄱ-ㅣ가-힣]+대로))")
    m = p.match(addr)
    if m is not None:
        result = re.sub(r'\d', '', m.group().split(' ')[1])
        if result == "검단동" or result == "가좌동":
            gu_name = "서구"
        elif result == "논현동":
            gu_name = "남동구"
        elif result == "계산동" or result == "효성동":
            gu_name = "계양구"
        elif result == "청천동":
            gu_name = "부평구"
        else:
            gu_name = "계양구"
        addr = addr.replace(" ", f" {gu_name} ")

    return addr
        
user_history.ADDR = user_history.ADDR.apply(add_gu)
user_history.head()
```

## IBCF(Item-based collaborative filtering)

```python
import numpy as np
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

전체 데이터로 행렬을 만들 경우 메모리 에러가 발생한다.
-> 2만개의 데이터만 분리하여 테스트

```python
ratings = cf_df.head(20000)
```

```python
ratings
```

```python
# train, test set 분리
x = ratings.copy()
y = ratings['USERID']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 12)
```

```python
review_count = x_train.groupby('USERID').count().sort_values('AVG(TOTALSCORE)', ascending = False)
review_count
```

```python
review_count[review_count['MENUID'] == 1]
```

```python
# full matrix 생성
# MENUID를 인덱스로 matrix 생성
rating_matrix = x_train.pivot(values = 'AVG(TOTALSCORE)', index = 'MENUID', columns = 'USERID')
```

```python
rating_matrix
```

```python
from sklearn.metrics import mean_absolute_error as mae

# RMSE 계산 
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2)) 

# def MAE(y_true, y_pred):
#     return 

def score(model, neighbor_size = 20): # 모델을 입력받음
    id_pairs = zip(x_test['USERID'], x_test['MENUID']) # test set의 userid와 menuid pair 
    y_pred = np.array([model(user, movie) for (user, movie) in id_pairs]) # 예측값 
    y_true = np.array(x_test['AVG(TOTALSCORE)']) # 실제값
    return RMSE(y_true, y_pred)
```

```python
# 아이템 pair의 Cosine similarities 계산
from sklearn.metrics.pairwise import cosine_similarity
matrix_dummy = rating_matrix.copy().fillna(0)
item_similarity = cosine_similarity(matrix_dummy, matrix_dummy)
item_similarity = pd.DataFrame(item_similarity, index = rating_matrix.index, columns = rating_matrix.index)
```

```python
item_similarity
```

```python
def ibcf(user_id, menu_id):
    if user_id in rating_matrix: # 사용자가 train set에 있는지 확인 
        if menu_id in rating_matrix: # 메뉴가 train set에 있는지 확인 
            # 현재 메뉴와 다른 메뉴의 similarity 값 가져오기
            sim_scores = item_similarity[menu_id]
            
            # 현 사용자의 모든 rating 값 가져오기
            user_rating = rating_matrix[user_id]
            
            # 사용자가 평가하지 않은 메뉴 index 가져오기
            non_rating_idx = user_rating[user_rating.isnull()].index
            
            # 사용자가 평가하지 않은 메뉴 제거
            user_rating = user_rating.dropna()
            
            # 사용자가 평가하지 않은 메뉴의 similarity 값 제거 
            sim_scores = sim_scores.drop(non_rating_idx)
            
            # 현 메뉴에 대한 사용자의 예상 rating 계산, 가중치는 현 메뉴와 사용자가 평가한 메뉴의 유사도
            mean_rating = np.dot(sim_scores, user_rating) / sim_scores.sum()
            
        else:
            mean_rating = 3.0
    else:
        mean_rating = 3.0
    return mean_rating
```

성능측정

```python
rmse, mae = score(ibcf)
print(f"RMSE: {rmse}\nMAE: {mae}")
```

## IBCF + KNN

```python
### KNN을 활용한 IBCF
def ibcf_knn(user_id, item_id, neighbor_size = 1):
    if user_id in rating_matrix: # user id가 train set에 있는지 
        if menu_id in rating_matrix: # menu id가 train set에 있는지
            # 유저가 먹은 다른 메뉴들의 평점
            choiced_menu_rating = rating_matrix[user_id].dropna()
            
            # choiced_menu_rating = choiced_menu['AVG(TOTALSCORE)']
            
            # 유저가 먹은 다른 메뉴들과 menu_id와의 유사도
            not_choiced = rating_matrix[user_id][rating_matrix[user_id].isnull()].index            
            sim = item_similarity[menu_id].drop(index = not_choiced)
            
            # neighbor size가 지정되지 않은 경우
            if neighbor_size == 0:
                mean_rating = np.dot(choiced_menu_rating, sim) / sim.sum()
            # neighbor size가 지정된 경우
            else:
                # 지정된 neighbor size 값과 사용자가 평가한 메뉴 수 중 작은 것으로 결정
                neighbor_size = min(neighbor_size, len(sim))
                print(neighbor_size)
                
                # array로 바꾸기(argsort를 사용하기 위함)
                sim = np.array(sim)
                choiced_menu_rating = np.array(choiced_menu_rating)
                
                # 유사도를 순서대로 정렬
                user_idx = np.argsort(sim)
                
                # 유사도를 neighbor size 만큼 받기
                sim = sim[user_idx][-neighbor_size:]
                
                # 메뉴 rating을 neighbor size 만큼 받기
                choiced_menu_rating = choiced_menu_rating[user_idx][-neighbor_size:]
                
                # 최종 예측값 계싼
                mean_rating = np.dot(sim, choidced_menu_rating) / sim.sum()
        else:
            mean_rating = 3.0
    else:
        mean_rating = 3.0
    return mean_rating
            
```
